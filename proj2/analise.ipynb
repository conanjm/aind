```python
SA = set(list(results[results.columns[1]].values))
HS = set(list(results[results.columns[2]].values))
```

```python
results_sep = []
i=0
for search in SA:
    for heuristic in HS:
        temp = results.loc[(results[results.columns[1]] == search) & (results[results.columns[2]] == heuristic)]
        if len(temp)>0:
            results_sep.append(temp)
            
results_sep
```

```python
for i in range(0,len(results_sep)):
    x = 3
    yy = [4, -1]

    for y in yy:
        plt.plot(results_sep[i][[results.columns[x]]].values, \
                 results_sep[i][[results.columns[y]]].values, \
                 'b--')
        plt.plot(results_sep[i][[results.columns[x]]].values, \
                 results_sep[i][[results.columns[y]]].values, \
                 's')
        plt.grid(True)
        plt.ylabel(results.columns[y])
        plt.xlabel(results.columns[x])
        plt.title(results.columns[y]+' vs '+results.columns[x])
        plt.draw()
        plt.savefig("fig/results_"+str(i)+str(x)+str(abs(y))+".png")
        plt.show()
```
